# Cardio Good Fitness Case Study - Descriptive Statistics
The market research team at AdRight is assigned the task to identify the profile of the typical customer for each treadmill product offered by CardioGood Fitness. The market research team decides to investigate whether there are differences across the product lines with respect to customer characteristics. The team decides to collect data on individuals who purchased a treadmill at a CardioGoodFitness retail store during the prior three months. The data are stored in the CardioGoodFitness.csv file.

### The team identifies the following customer variables to study: 
  - product purchased, TM195, TM498, or TM798; 
  - gender; 
  - age, in years; 
  - education, in years; 
  - relationship status, single or partnered; 
  - annual household income ; 
  - average number of times the customer plans to use the treadmill each week; 
  - average number of miles the customer expects to walk/run each week; 
  - and self-rated fitness on an 1-to-5 scale, where 1 is poor shape and 5 is excellent shape.

### Correlation Matrix

```python
# Creating a copy of dataset.

mydata_copy = mydata
```

```python
# Using one hot encoding on the categorical dataset.

mydata_copy = pd.get_dummies(mydata_copy, columns=['Product'], prefix='Product')
mydata_copy = pd.get_dummies(mydata_copy, columns=['Gender'], prefix='Gender')
mydata_copy = pd.get_dummies(mydata_copy, columns=['MaritalStatus'], prefix='Status')
```

```python
plt.figure(figsize=(12, 9))
sns.heatmap(mydata_copy.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**Observations**

1. Strong Correlations:

- Miles and Usage (0.76): There is a strong positive correlation between the number of miles a customer runs and their usage of the products. This suggests that customers who use the products more frequently tend to run more miles.
- Usage and Fitness (0.67): Customers who use the products more also have higher fitness levels, indicating that product usage is closely linked to physical activity.
- Fitness and Miles (0.79): There is a strong positive correlation between fitness levels and miles run, showing that fitter customers tend to run more miles.
- Income and Education (0.63): Higher income is correlated with higher levels of education among customers.

2. Moderate Correlations:
- Fitness and Product_TM798 (0.71): Customers who are fitter tend to use Product_TM798 more frequently.
- Usage and Product_TM798 (0.65): Higher usage of products is moderately correlated with the use of Product_TM798.

3. Negative Correlations:
- Product_TM195 and Product_TM798 (-0.48): Customers who use Product_TM195 tend to use Product_TM798 less frequently and vice versa.
- Product_TM195 and Miles (-0.35): There is a moderate negative correlation between the use of Product_TM195 and the number of miles run, indicating that customers using  Product_TM195 may be less active in terms of running miles.
